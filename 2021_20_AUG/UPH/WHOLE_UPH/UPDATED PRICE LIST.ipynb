## JOINING TWO EXCEL FILES

```python
import pandas as pd

df1=pd.read_excel("UPH_NEWUPDATE_CDN.xlsx")
df2=pd.read_excel("UPH_NEWUPDATE_US.xlsx")

values1=df1[["PriceList", "Catalog", "GradeGroup", "Price"]]
values2=df2[["PriceList", "Catalog", "GradeGroup", "Price"]]

dataframes=[values1,values2]
join=pd.concat(dataframes)

join.to_excel("UPH_NEWUPDATE.xlsx")
join

```

## NOW WE WANT TO REMOVE THE DUPLICATED ROWS:

```python
join=join.drop_duplicates(subset=['PriceList','Catalog','GradeGroup'])
join.to_excel('UPH_Aug_20.xlsx')
print(len(join['Catalog']))
```

## Now we need to update Aug 6 file according to Aug 20 (2021)

```python
Aug_20=pd.read_excel('UPH_Aug_20.xlsx')
Aug_6=pd.read_excel('UPH_Aug_6.xlsx')
print(len(Aug_20))
print(len(Aug_6))
```

## First we create a list for all three columns for two different Excel files:

```python
PriceList_Aug6=[]
Catalog_Aug6=[]
GradeGroup_Aug6=[]

PriceList_Aug20=[]
Catalog_Aug20=[]
GradeGroup_Aug20=[]

Price_Aug6=[]

for i in Aug_6['PriceList']:
    PriceList_Aug6.append(i)
    
for i in Aug_6['Catalog']:
    Catalog_Aug6.append(i)
    
for i in Aug_6['GradeGroup']:
    GradeGroup_Aug6.append(i)

for i in Aug_20['PriceList']:
    PriceList_Aug20.append(i)
    
for i in Aug_20['Catalog']:
    Catalog_Aug20.append(i)
    
for i in Aug_20['GradeGroup']:
    GradeGroup_Aug20.append(i)

print(len(PriceList_Aug6))
print(len(Catalog_Aug6))
print(len(GradeGroup_Aug6))

    
    
print(len(PriceList_Aug20))
print(len(Catalog_Aug20))
print(len(GradeGroup_Aug20)) 


a=[PriceList_Aug6,Catalog_Aug6,GradeGroup_Aug6]
b=[PriceList_Aug20,Catalog_Aug20,GradeGroup_Aug20]

length_Aug6=len(a[0])
length_Aug20=len(b[0])

print("This is the length of Aug6: ")
print(length_Aug6)
print("This is the length of Aug20: ")
print(length_Aug20)


for j in range(0,length_Aug6,1):
    print(a[0][j])
    





```

```python
for k in range(0,length_Aug20,1):
        for j in range(0,length_Aug6,1):
            if (a[0][j]==b[0][k] and a[1][j]==b[1][k] and a[2][j]==b[2][k]):
                Aug_6.at[j,'Price']=Aug_20.at[k,'Price']
                
for i in Aug_6['Price']:
    Price_Aug6.append(i)
    
    
    
print(Aug_6['Price'])
print(len(Aug_6['Price']))
```

```python
from itertools import zip_longest
import csv
Table=[PriceList_Aug6,Catalog_Aug6,GradeGroup_Aug6,Price_Aug6]
export_data = zip_longest(*Table, fillvalue = '')
with open('Check.csv', 'w', encoding="ISO-8859-1", newline='') as myfile:
      wr = csv.writer(myfile)
      wr.writerow(("PriceList", "Catalog", "GradeGroup", "Price"))
      wr.writerows(export_data)
myfile.close()





```
