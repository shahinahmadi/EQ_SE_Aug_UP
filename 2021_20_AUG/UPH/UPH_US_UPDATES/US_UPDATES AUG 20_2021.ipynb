```python
import numpy
import csv
import pandas as pd
import numpy as np
from itertools import zip_longest




US_data=pd.read_excel("UPH_NEWEST.xlsx", sheet_name='USD')
US_data['ITEM_F'].replace('', np.nan, inplace=True)
US_data.dropna(subset = ["ITEM_F"], inplace=True)

    
      
US_data.to_csv('US_landed1.csv', header=False, index=False)  
US_data1= pd.read_csv("US_landed1.csv")
US_data2=pd.read_csv("US_landed1.csv", 
names=['ITEM_F', 'ITEM_L', 'Grade10','Grade20','Grade30','Grade40','Grade50','Grade60','Grade70','GradeF','GradeI','Grade41', 
'Grade42','Grade100','Grade200','Grade300','Grade400','Grade500','Grade600','Grade700', 'Grade800','GradeM'])
pd.set_option("max_rows", None)
US_data2
print(len(US_data2["ITEM_F"]))


length1=len(US_data2['ITEM_F'])
LPriceListName=[]
LCatalog=[]
LGradeGroup=[]
LPrice=[]
GradeNumber_Fabric=[10,20,30,40,50,60,70,'F','I',41,42]
GradeNumber_Leather=[100,200,300,400,500,600,700,800,'M']




print("------------------------------------------Organizng the data----------------------------------------------")

print("************************************Fabric*********************************************")


for i in US_data2["ITEM_F"]:
    for j in numpy.arange(1,12,1):
        LPriceListName.append('UFMX 1711E V5')
        
    

for i in US_data2["ITEM_F"]:
      for j in numpy.arange(1,12, 1):
                 LCatalog.append(i) 


for i in US_data2["ITEM_F"]:
    for j in GradeNumber_Fabric:
           LGradeGroup.append(j)

for i in range(0,length1,1):
    LPrice.append(US_data2.Grade10[i])
    LPrice.append(US_data2.Grade20[i])
    LPrice.append(US_data2.Grade30[i])
    LPrice.append(US_data2.Grade40[i])
    LPrice.append(US_data2.Grade50[i])
    LPrice.append(US_data2.Grade60[i])
    LPrice.append(US_data2.Grade70[i])
    LPrice.append(float(US_data2.GradeF[i]))
    LPrice.append(float(US_data2.GradeI[i]))
    LPrice.append(US_data2.Grade41[i])
    LPrice.append(US_data2.Grade42[i])
    

print("***********************************Leather*************************************************")
US_data_L=pd.read_excel("UPH_NEWEST.xlsx", sheet_name='USD')
US_data_L['ITEM_L'].replace('', np.nan, inplace=True)
US_data_L.dropna(subset = ["ITEM_L"], inplace=True)

    
      
US_data_L.to_csv('US_landed2.csv', header=False, index=False)  
US_data3= pd.read_csv("US_landed2.csv")
US_data4=pd.read_csv("US_landed2.csv", 
names=['ITEM_F', 'ITEM_L', 'Grade10','Grade20','Grade30','Grade40','Grade50','Grade60','Grade70','GradeF','GradeI','Grade41', 
'Grade42','Grade100','Grade200','Grade300','Grade400','Grade500','Grade600','Grade700', 'Grade800','GradeM'])



length2=len(US_data4['ITEM_L'])


for i in US_data4["ITEM_L"]:
    for j in numpy.arange(1,10,1):
        LPriceListName.append("UFMX 1711E V5")


    
for i in US_data4["ITEM_L"]:
      for j in numpy.arange(1,10, 1):
                 LCatalog.append(i) 
                
                
for i in US_data4["ITEM_L"]:
    for j in GradeNumber_Leather:
           LGradeGroup.append(j)
            
            
for i in range(0,length2,1):
    LPrice.append(US_data4.Grade100[i])
    LPrice.append(US_data4.Grade200[i])
    LPrice.append(US_data4.Grade300[i])
    LPrice.append(US_data4.Grade400[i])
    LPrice.append(US_data4.Grade500[i])
    LPrice.append(US_data4.Grade600[i])
    LPrice.append(US_data4.Grade700[i])
    LPrice.append(US_data4.Grade800[i])
    LPrice.append(float(US_data4.GradeM[i]))

Table=[LPriceListName,LCatalog,LGradeGroup,LPrice]
export_data = zip_longest(*Table, fillvalue = '')
with open('Upload_L.csv', 'w', encoding="ISO-8859-1", newline='') as myfile:
      wr = csv.writer(myfile)
      wr.writerow(("PriceList", "Catalog", "GradeGroup", "Price"))
      wr.writerows(export_data)
myfile.close()

Final_data=pd.read_csv('Upload_L.csv')
Final_data['Price'].replace('', np.nan, inplace=True)
Final_data.dropna(subset = ["Price"], inplace=True)
#writer=ExcelWriter("Upload_L_X.xlsx")
#df1.to_excel(writer,sheet_name='Sheet_L', index=False)

print("------------------------------End of data organization---------------------------------------------")
pd.set_option("max_rows", None)
Final_data.to_excel('UPH_NEWUPDATE_US.xlsx', header=True, index=False)
Final_data
    


    
    

    







```
