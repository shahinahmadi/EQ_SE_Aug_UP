```python
import pandas as pd


CDN_data=pd.read_excel("old_data.xlsx", sheet_name='UPH_CDN_OLD')
US_data=pd.read_excel("old_data.xlsx", sheet_name='UPH_US_OLD')


values1=CDN_data[["PriceList", "Catalog", "GradeGroup", "Price"]]
values2=US_data[["PriceList", "Catalog", "GradeGroup", "Price"]]

dataframes=[values1,values2]
join=pd.concat(dataframes)

#Saving the joint old file to one file named OLD_UPH.xlsx
join.to_excel("OLD_UPH.xlsx")
#Reading the old file
Aug_Old=pd.read_excel("OLD_UPH.xlsx")
#Reading the Aug 20 price list file
Aug_20=pd.read_excel("UPH_Aug_20.xlsx")

print(len(Aug_Old))
print(len(Aug_20))
```

```python
PriceList_Aug20=[]
Catalog_Aug20=[]
GradeGroup_Aug20=[]

Price_Aug20=[]


PriceList_Aug_Old=[]
Catalog_Aug_Old=[]
GradeGroup_Aug_Old=[]


for i in Aug_20['PriceList']:
    PriceList_Aug20.append(i)
    
for i in Aug_20['Catalog']:
    Catalog_Aug20.append(i)
    
for i in Aug_20['GradeGroup']:
    GradeGroup_Aug20.append(i)
    
    
for i in Aug_Old['PriceList']:
    PriceList_Aug_Old.append(i)
    
for i in Aug_Old['Catalog']:
    Catalog_Aug_Old.append(i)
    
for i in Aug_Old['GradeGroup']:
    GradeGroup_Aug_Old.append(i)


    
print(len(PriceList_Aug20))
print(len(Catalog_Aug20))
print(len(GradeGroup_Aug20))
    
    

print(len(PriceList_Aug_Old))
print(len(Catalog_Aug_Old))
print(len(GradeGroup_Aug_Old))

    
 
a=[PriceList_Aug20,Catalog_Aug20,GradeGroup_Aug20]
b=[PriceList_Aug_Old,Catalog_Aug_Old,GradeGroup_Aug_Old]

length_Aug20=len(a[1])
length_Aug_Old=len(b[1])


print("This is the length of Aug20: ")
print(length_Aug20)

print("This is the length of Aug_Old: ")
print(length_Aug_Old)





for j in range(0,length_Aug20,1):
    print(a[0][j])
    print(b[0][j])
    print(a[1][j])
    print(b[1][j])
    print(a[2][j])
    print(b[2][j])
    





```

## We replace the Aug 20 data with the old data:

```python
for k in range(0,length_Aug_Old,1):
        for j in range(0,length_Aug20,1):
            if (a[0][j]==b[0][k] and a[1][j]==b[1][k] and a[2][j]==b[2][k]):
                Aug_20.at[j,'Price']= Aug_Old.at[k,'Price']
                print(Aug_Old.at[k,'Price'])
               

for i in Aug_20['Price']:
    Price_Aug20.append(i)

    
print(Price_Aug20)   
print(Aug_20['Price'])
print(len(Aug_20['Price']))
```

## Now we save the list to a csv file as below:

```python
from itertools import zip_longest
import csv
Table=[PriceList_Aug20,Catalog_Aug20,GradeGroup_Aug20,Price_Aug20]
export_data = zip_longest(*Table, fillvalue = '')
with open('Check.csv', 'w', encoding="ISO-8859-1", newline='') as myfile:
      wr = csv.writer(myfile)
      wr.writerow(("PriceList", "Catalog", "GradeGroup", "Price"))
      wr.writerows(export_data)
myfile.close()
```
